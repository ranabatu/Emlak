```python
# sıralı ilişkiler için label encoder kullanılır
label_encoder = LabelEncoder()

# Dönüştürmek istediğimiz sütunların listesi
label_kolonlar = ['isinma', 'tip', 'ilce', 'mahalle']

for col in label_kolonlar:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
# 1. Çeyreklik (Q1) ve 3. Çeyreklik (Q3) hesaplama
Q1 = df['fiyat'].quantile(0.25)
Q3 = df['fiyat'].quantile(0.75)
IQR = Q3 - Q1  # IQR hesaplama

# Aykırı değer sınırlarını belirleme
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri filtrelemek yerine onları veri çerçevesinden silme
df = df[(df['fiyat'] >= lower_bound) & (df['fiyat'] <= upper_bound)]

# Temizlenmiş veri çerçevesini kontrol etme
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Belirli sütunlara MinMaxScaler uygulamak istediğiniz sütunlar
minmax_columns = ['net_mt', 'brut_mt', 'oda', 'banyo', 'salon', 'fiyat']

# Belirli sütunlara StandardScaler uygulamak istediğiniz sütunlar
standard_columns = ['kat', 'bina_kat', 'yas']

# MinMaxScaler ve StandardScaler nesnelerini oluşturma
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# MinMaxScaler uygulanacak sütunlar için ölçekleme
df[minmax_columns] = minmax_scaler.fit_transform(df[minmax_columns])

# StandardScaler uygulanacak sütunlar için ölçekleme
df[standard_columns] = standard_scaler.fit_transform(df[standard_columns])

# Sonuçları kontrol etme
print(df.head())
```

```python
# Verisetini bağımlı ve bağımsız değişkenler olarak ayırıyoruz
x=df.drop(['fiyat'],axis=1)
y=df['fiyat']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Modeli oluşturma
model = LinearRegression()

# Modeli eğitme
model.fit(x_train, y_train)
```

```python
# Tahmin yapma
y_pred = model.predict(x_test)

# Modelin performansını değerlendirme
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
# İlk 5 tahmini ve gerçek değerleri karşılaştırma
print("Tahminler:", y_pred[:5])
print("Gerçek Değerler:", y_test[:5].values)
```
